```python
import pandas as pd
from IPython.display import display, Markdown

display(Markdown("## Load and Explore the Dataset"))
display(Markdown("**Date:** 2026-05-31"))
display(Markdown("**Student ID:** vingut8011"))

data = pd.read_csv(
    r'C:\Users\iaasi\Documents\SDC485DocumentsECPI_AI_Class\winequality-red.csv'
)

print("Dataset Loaded Successfully!")
print("Dataset Shape:", data.shape)

display(data.head())
data.info()
display(data.describe())
```

```python
from sklearn.preprocessing import StandardScaler

display(Markdown("## Preprocess the Data"))
display(Markdown("**Date:** 2026-05-31"))
display(Markdown("**Student ID:** vingut8011"))

print("Missing Values Before:")
print(data.isnull().sum())

print("\nDuplicate Rows Before:", data.duplicated().sum())

data = data.drop_duplicates()

print("Duplicate Rows After:", data.duplicated().sum())

# Separate features from quality column
X = data.drop('quality', axis=1)

# Standardize numerical features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("\nPreprocessing Complete")
print("Original Data Shape:", data.shape)
print("Scaled Feature Shape:", X_scaled.shape)
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

display(Markdown("## Dendrogram with Hierarchical Clustering"))
display(Markdown("**Date:** 2026-05-31"))
display(Markdown("**Student ID:** vingut8011"))

# Use hierarchical clustering
linked = linkage(X_scaled, method='ward')

# Plot dendrogram
plt.figure(figsize=(12, 6))
dendrogram(
    linked,
    truncate_mode='level',
    p=5,
    leaf_rotation=90,
    leaf_font_size=10
)

plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

```python
from sklearn.cluster import KMeans
from IPython.display import display, Markdown

display(Markdown("## Implement K-Means Clustering"))
display(Markdown("**Date:** 2026-05-31"))
display(Markdown("**Student ID:** vingut8011"))

# Initial cluster count from dendrogram
kmeans = KMeans(n_clusters=4, random_state=42, n_init=10)

# Fit model
clusters = kmeans.fit_predict(X_scaled)

print("Number of Clusters:", 4)
print("Cluster Labels:")
print(clusters[:20])
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

display(Markdown("## Elbow Method"))
display(Markdown("**Date:** 2026-05-31"))
display(Markdown("**Student ID:** vingut8011"))

inertia = []

for k in range(1, 11):
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    model.fit(X_scaled)
    inertia.append(model.inertia_)

plt.figure(figsize=(8,5))
plt.plot(range(1,11), inertia, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

display(Markdown("## Silhouette Scores"))
display(Markdown("**Date:** 2026-05-31"))
display(Markdown("**Student ID:** vingut8011"))

scores = []

for k in range(2, 11):
    model = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = model.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    scores.append(score)

plt.figure(figsize=(8,5))
plt.plot(range(2,11), scores, marker='o')
plt.title("Silhouette Scores")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Silhouette Score")
plt.grid(True)
plt.show()

for k, score in zip(range(2,11), scores):
    print(f"k={k}: {score:.4f}")
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from IPython.display import display, Markdown

display(Markdown("## Evaluate the Clustering"))
display(Markdown("**Date:** 2026-05-31"))
display(Markdown("**Student ID:** vingut8011"))

optimal_k = 2

kmeans_final = KMeans(
    n_clusters=optimal_k,
    random_state=42,
    n_init=10
)

cluster_labels = kmeans_final.fit_predict(X_scaled)

final_score = silhouette_score(X_scaled, cluster_labels)

print("Optimal Number of Clusters:", optimal_k)
print("Final Silhouette Score:", round(final_score,4))
```

```python
import matplotlib.pyplot as plt
from IPython.display import display, Markdown

display(Markdown("## Cluster Visualization"))
display(Markdown("**Date:** 2026-05-31"))
display(Markdown("**Student ID:** vingut8011"))

plt.figure(figsize=(8,6))

plt.scatter(
    X_scaled[:,0],
    X_scaled[:,1],
    c=cluster_labels,
    cmap='viridis'
)

plt.title("K-Means Clustering Results")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.colorbar(label="Cluster")

plt.show()
```
